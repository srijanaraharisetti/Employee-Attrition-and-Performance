## Outliers

```python
df.boxplot(column=["MonthlyIncome"])
props=dict(boxes='yellow', whiskers='green', medians='black', caps='blue')
ax=df['MonthlyIncome'].plot.box(color=props, patch_artist=True)
```

```python
## Treating outliers
q1= df['MonthlyIncome'].quantile(0.25)
q3= df['MonthlyIncome'].quantile(0.75)
print("Q1 = ", q1)
print("Q3 = ", q3)
```

```python
iqr = q3-q1
lb = q1-1.5*iqr
print("Lower Bound = ", lb)
ub = q3 + 1.5*iqr
print("Upper Bound = ", ub)
```

```python
df['MonthlyIncome'][df['MonthlyIncome'] < lb] = lb
df['MonthlyIncome'][df['MonthlyIncome'] > ub] = ub
```

```python
df.boxplot(column=["MonthlyIncome"])
props=dict(boxes='yellow', whiskers='green', medians='black', caps='blue')
ax=df['MonthlyIncome'].plot.box(color=props, patch_artist=True)
```

## Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
```

```python
rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5)
feature_selector = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=1)
```

```python
rf.fit(x,y)
```

```python
x_ = x.to_numpy()
```

```python
feature_selector.fit(x_,y)
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Attrition',axis=1),
df['Attrition'], test_size=0.30,
random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test,predictions))
```
